# Calculus

## Differentiation

### Chain rule

$$
\frac{dy}{dx} = \frac{dy}{du} \times \frac{du}{dx}
\newline
$$

$$
\textrm{If }\;
y = f(g(x)) \;
\textrm{then }\;
\frac{dy}{dx}=f'(g(x))g'(x)
$$

$$
\textrm{where y is a function of u, and u is another function of x}
$$

### Product rule

$$
\textrm{If }\;
y = uv \;
\textrm{then }\;
\frac{dy}{dx} = u\frac{dv}{dx} + v\frac{du}{dx}
$$


$$
\textrm{If }\;
f(x) = g(x)h(x) \;
\textrm{then }\;
f'(x)=g(x)h'(x) + h(x)g'(x)
$$

### Quotient rule

$$
\textrm{If }\;
y = \frac{u}{v} \;
\textrm{then }\;
\frac{dy}{dx} = \frac{v\frac{du}{dx} - u\frac{dv}{dx}}{v^{2}}
$$


$$
\textrm{If }\;
f(x) = \frac{g(x)}{h(x)} \;
\textrm{then }\;
f'(x)=\frac{h(x)g'(x) - g(x)h'(x)}{(h(x))^{2}}
$$

### Parametric differentiation

If $x$ and $y$ are given as functions of another parameter, $t$, then (from the chain rule):

$$
\frac{dy}{dx} = \frac{\frac{dy}{dt}}{\frac{dx}{dt}}
$$

### Exponentials & logarithms

$$
\textrm{If }\;
y = e^{kx} \;
\textrm{then }\;
\frac{dy}{dx}=ke^{kx}
$$

$$
\textrm{If }\;
y = \ln{x} \;
\textrm{then }\;
\frac{dy}{dx}=\frac{1}{x}
$$

$$
\textrm{If }\;
y = a^{kx} \;
\textrm{then }\;
\frac{dy}{dx}=a^{kx} k\ln a
$$

$$
\textrm{where k is a real constant and a>0}
$$


## Integration

$$
\int x^{n} \, dx = \frac{x^{(n+1)}}{n+1} + c
$$

$$
\int e^{x} \, dx = e^{x} + c
$$

$$
\int \frac{1}{x} \, dx = ln \lvert x \rvert + c
$$
